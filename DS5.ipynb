```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Load the dataset
file_path = 'D:/Data science ass/5/EDA1/Cardiotocographic.csv'
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Handle missing values (e.g., fill with the mean of each column or drop rows/columns with missing values)
df.fillna(df.mean(), inplace=True)

# Convert numerical columns to appropriate data types if necessary
df = df.apply(pd.to_numeric, errors='coerce')
```

```python
# Generate descriptive statistics
summary = df.describe()
print(summary)
```

```python
# Plot histograms for each column
df.hist(bins=30, figsize=(20, 15))
plt.show()
```

```python
# Plot boxplots for each column
df.plot(kind='box', subplots=True, layout=(6, 4), figsize=(20, 15), sharex=False, sharey=False)
plt.show()
```

```python
# Calculate the correlation between LB and AC
correlation = df['LB'].corr(df['AC'])
print(f'Correlation between LB and AC: {correlation}')
```
